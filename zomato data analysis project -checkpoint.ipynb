# ZOMATO DATA ANALYSIS PROJECT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Zomato data 1.csv')
print(df.head())
```

```python
df
```

# convert the datatype of column rate..

```python
def handlerate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)
    
df['rate']=df['rate'].apply(handlerate)
print(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#  what type of restaurant do the majority of customers order from..
```

```python
df.head(3)
```

```python
sns.countplot(x=df['listed_in(type)'],data=df,hue='listed_in(type)',palette='viridis',edgecolor='black')
plt.xlabel("type of resturant")
plt.show()
```

```python
#  conclusion: majority of customers order from the dining resturant.
```

```python
# ques 2.. how many votes has each type of resturant received from customers..
```

```python
grouped_data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes': grouped_data})
plt.plot(result,c='green',marker='o')
plt.xlabel('type of resturant',c='blue',size=20)
plt.ylabel('votes',c='blue',size=20)
plt.show()
```

```python
# conclusion.. dining resturant has received maximum votes.
```

```python
# ques...3... what are the ratings that the majority of resturants have received...
```

```python
plt.hist(df['rate'],bins =5,color='skyblue',edgecolor='black')
plt.title('ratings distribution')
plt.show()
```

```python
# conclusion: the majority resturant receieved ratings from 3.5 to 4.
```

```python
# ques  4... zomato has observed that most couples order most of their food online. what is the avg. spending on each order..
```

```python
df.head()
```

```python
couple_data=df['approx_cost(for two people)']
sns.countplot(x=couple_data, data=df,hue='approx_cost(for two people)',palette='Set2',edgecolor='black')
plt.legend([],[],frameon=False)
plt.show()
```

```python
#  conclusion...  the majority of couples prefer resturants with an approx cost of 300 rupees....
```

```python
# ques 5...which mode(online or offline) has received the maximum ratings...
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate', data=df,color='lightblue')
plt.show()
```

```python
#  conclusion... offline order received lower ratings in comparison to online order.
```

```python
# ques5... which type of resturant received more offline orders,so that zomato can replace customers with some good offers..
```

```python
pivot_table=df.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='Pastel1',fmt='d')
plt.title('heatmap')
plt.xlabel('online order')
plt.ylabel('Listed in (type)')
plt.show()
```

```python
#  conclusion... dining resturants primiarly accept offline orders,whereas cafes primiarily receieve online oredrs. this suggests taht 
# clients preffer orders in person at resturants, but prefer online ordering at cafes..
```
